```python
import requests
```

```python
link = "https://www.goodreads.com/choiceawards/best-fiction-books-2023"
```

```python
response = requests.get(link)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'lxml')
```

```python
arr = []
for link_element in soup.find_all('a'):
    href = link_element.get('href')

    if "/book/show" in str(href):
        arr.append(href)
```

```python
arr = list(set(arr))
```

```python
all_links = [f"https://www.goodreads.com{i}" for i in arr]
```

```python
all_links
```

```python
books = []
for link in all_links:
    book_response = requests.get(link)
    soup = BeautifulSoup(book_response.text, 'lxml')

    cover = soup.find("div", class_="BookPage__bookCover")
    image_tag = cover.find('img')

    # Extract the src attribute
    if image_tag:
        image_src = image_tag['src']
        print(image_src)
    else:
        print("Image tag not found")
    image_src
    book_info = soup.find("div", class_="BookPage__mainContent")
    title = soup.find('h1', class_='Text Text__title1').text.strip()
    author_link = soup.find('a', class_='ContributorLink')
    author = author_link.span.text.strip()

    genres = []
    for genre_button in soup.find_all('a', class_='Button Button--tag-inline Button--small'):
        genres.append(genre_button.text.strip())


    books.append({
        'link': link,
        'title': title,
        'image': image_src,
        'author': author,
        'genres': genres
    })
```

```python
chapters = []
num = 1
while True:
    if num==3: 
        link = "https://lythrumpress.com.au/chapter/yellowface-book-chapter-no-3/"
    else:
        link = f"https://lythrumpress.com.au/chapter/yellowface-chapter-no-{num}/"
    print(num)
    num += 1
    try:
        response = requests.get(link)
    except Exception as e:
        print(e)
        break

    soup = BeautifulSoup(response.text, "lxml")
    r = soup.find_all("p")
    title = soup.title
    paras = [p.text for p in r]
    texts = "\n".join(paras)
    if texts == "Read Free Novel Online\xa0":
        print(link)
        break

    #
    chapters.append(
        {   "id": num,
            "chapter_name": title,
            "chapter_text": texts,
        }
    )
```

```python
for i in chapters:
    i["chapter_name"] = i["chapter_name"].text
```

```python
import json
json.dump(chapters, open("Yellowface by R.F. Kuang Chapter.json", "w"), indent=4)
```
